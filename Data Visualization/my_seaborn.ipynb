```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_titanic = pd.read_csv('titanic.csv')
```

```python
sns.displot(data=df_titanic,
            x='Age', # ใส่ชื่อ column ลงในแกนได้เลย
            kde=True, # เส้นโค้ง
            alpha=0.5, # ความจาง
            linewidth=0.5) # ความคม/เข้ม ของขอบ
```

```python
sns.displot(data=df_titanic,
            x='Age', 
            kde=True, 
            alpha=0.4, 
            linewidth=0,
            hue='Sex',  # แบ่งตามเพศ
            palette=['r', 'b']) # สี
```

```python
sns.displot(data=df_titanic,
            x='Age', 
            kde=True, 
            alpha=0.4, 
            linewidth=0,
            hue='Pclass', # แบ่งตาม Pclass
            palette=sns.color_palette("colorblind", 3))

# ใช้ matplotlib ได้
plt.xlabel('Passenger Age')
plt.ylabel('Frequency')
plt.title('Passenger age & Frequency')
```

```python
sns.displot(data=df_titanic,
            x='Age', 
            kind='kde', # เอาแค่ kde
            alpha=1, 
            linewidth=1,
            hue='Pclass',
            palette=sns.color_palette("colorblind", 3))
```

```python
sns.displot(data=df_titanic,
            x='Age',
            kind='kde',
            hue='Pclass',
            alpha=0.7,
            linewidth=1.5,
            palette=['r', 'g', 'b'])

plt.xlabel('Passenger age')
plt.ylabel('Distributions of age according to Pclass')

age_means = df_titanic.groupby('Pclass').Age.mean()
# ax vertical line (plot เส้น ค่าเฉลี่ยเอง)
plt.axvline(x=age_means[1], c='r', ls=':')
plt.axvline(x=age_means[2], c='g', ls='--')
plt.axvline(x=age_means[3], c='b', ls='-.')
```
