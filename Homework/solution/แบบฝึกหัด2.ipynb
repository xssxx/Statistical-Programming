- หากในคอลัมน์ Cabin เป็นค่า "C23 C25 C27" ในคอลัมน์ที่สร้างจะบันทึก "C" 
- หากในคอลัมน์ Cabin เป็นค่า "B58 B60 C2" ในคอลัมน์ที่สร้างจะบันทึก "B"

```python
df = pd.read_csv("/content/drive/MyDrive/KU/0.Teaching/StatComp/2565_1/Basic_Python/BasicPython/titanic.csv")
df.head()
```

นิสิตแสดงวิธีสร้างคอลัมน์

```python
#your code
df['CabinLetter'] = [s[0] if not pd.isnull(s) else np.NaN for s in df.Cabin]
df.head(20)
```

```python
#หรือ
df['CabinLetter'] = df.Cabin.str.get(0)
df.head(20)
```

1. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารกี่คน และมีจำนวน NaN เท่าไหร่

```python
#your code
df.CabinLetter.value_counts(dropna=False)
```

2. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารที่รอดและเสียชีวิตอย่างละเท่าไหร่ cabin ใดมีจำนวนผู้เสียชีวิตมากที่สุด

```python
#your code
letterCabin_survive = pd.crosstab(df.CabinLetter, df.Survived, margins=False)
letterCabin_survive.sort_values(by=0, ascending=False)
```

```python
letterCabin_survive.index[np.argmax(letterCabin_survive[0])]
```

3. cabin ใดมีค่า ratio ผู้รอดชีวิต / ผู้เสียชีวิต สูงสุด

```python
#your code
letterCabin_survive['ratio'] = letterCabin_survive[1]/letterCabin_survive[0]
letterCabin_survive.sort_values(by='ratio', ascending=False)
```

```python
letterCabin_survive.index[np.argmax(letterCabin_survive.ratio)]
```

3. cabin ใดมีค่าเฉลี่ยราคาตั๋ว (Fare) สูงสุด

```python
#your code
df.filter(['Fare','CabinLetter']).groupby(['CabinLetter']).mean().sort_values(by='Fare', ascending=False)
```

4. จงแสดงค่าเฉลี่ยราคาตั๋ว (Fare) ของผู้รอดชีวิต และผู้เสียชีวิต ในแต่ละ cabin หากเปรียบเทียบราคาเฉลี่ยของผู้ที่รอดและเสียชีวิตจะพบว่า

```python
#your code
df.filter(['Survived','Fare','CabinLetter']).groupby(['CabinLetter', 'Survived']).mean()
```
