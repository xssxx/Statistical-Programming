- หากในคอลัมน์ Cabin เป็นค่า "C23 C25 C27" ในคอลัมน์ที่สร้างจะบันทึก "C" 
- หากในคอลัมน์ Cabin เป็นค่า "B58 B60 C2" ในคอลัมน์ที่สร้างจะบันทึก "B"

นิสิตแสดงวิธีสร้างคอลัมน์

```python
#your code
df["CabinLetter"] = df["Cabin"].str[0]
df.head()
```

1. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารกี่คน และมีจำนวน NaN เท่าไหร่

```python
#your code
counts = df["CabinLetter"].value_counts(dropna=False).sort_index()
counts
```

2. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารที่รอดและเสียชีวิตอย่างละเท่าไหร่ cabin ใดมีจำนวนผู้เสียชีวิตมากที่สุด

```python
#your code
ans = pd.crosstab(df['CabinLetter'], df['Survived'])
print(ans )
print("Cabin C มีผู้เสียชีวิตมากที่สุด")
```

3. cabin ใดมีค่า ratio ผู้รอดชีวิต / ผู้เสียชีวิต สูงสุด

```python
#your code
max_value = 0
max_key = ''

for key, value in pd.crosstab(df.Survived, df.CabinLetter).dropna().items():
    if value[1] / value[0] > max_value:
        max_value = value[1] / value[0]
        max_key = key
    print("Cabin {} = {:.2f}".format(key, value[1] / value[0]))

print(f"Cabin {max_key} มีค่า ratio ผู้รอดชีวิต / ผู้เสียชีวิต สูงสุด")
```

3. cabin ใดมีค่าเฉลี่ยราคาตั๋ว (Fare) สูงสุด

```python
#your code
print("Cabin ที่มีค่าราคาเฉลี่ยตั๋วสูงสุดคือ")
ans = df.filter(["Cabin", "Fare"]).groupby(["Cabin"]).mean()
print(ans.idxmax())
```

4. จงแสดงค่าเฉลี่ยราคาตั๋ว (Fare) ของผู้รอดชีวิต และผู้เสียชีวิต ในแต่ละ cabin หากเปรียบเทียบราคาเฉลี่ยของผู้ที่รอดและเสียชีวิตจะพบว่า

```python
#your code
grouped = df.groupby(['Cabin', 'Survived'])
avg_fare = grouped['Fare'].mean()
print(avg_fare)
```
