3.2 สุ่มตัวอย่างนักศึกษามาจำนวนหนึ่ง พบว่าคะแนนสอบวิชาสถิติ เป็นดังตาราง 'ex3_2_score.csv' จงหาค่า

```python
df_score = pd.read_csv('/content/drive/MyDrive/1st Year 1st Semester/Statistical Programming/ex3_2_score - ex3_2_score.csv')
df_score.head(8)
```

```python
lower = df_score["score"].str.split("-").str[0].astype(int)
upper = df_score["score"].str.split("-").str[1].astype(int)
df_score["mean_score"] = (upper + lower) / 2
df_score["mean_student"] = (df_score["mean_score"] * df_score["num_student"]).sum() / df_score["num_student"].sum()
print(df_score["num_student"].sum())
df_score.head(8)
```

**3.2.1 ค่าเฉลี่ย**

```python
#your code
mean = df_score["mean_student"].sum()
mean
```

**3.2.2 ค่ามัธยฐาน**

```python
#your code
median = df_score["num_student"].median()
median
```

**3.2.3 ค่าฐานนิยม**

```python
#your code
mode = df_score["num_student"].mode()
mode
```

**3.2.4 Q<sub>1</sub> และ Q<sub>3</sub>**

```python
#your code
Q1 = df_score["num_student"].quantile(1/4) # Quartile 1
Q3 = df_score["num_student"].quantile(3/4) # Quartile 3
print(f"Q1 = {Q1}")
print(f"Q3 = {Q3}")
```

**3.2.5 D<sub>7</sub>**

```python
#your code
D7 = df_score["num_student"].quantile(7/10) # Decile 7
D7
```

**3.2.6 P<sub>75</sub>**

```python
#your code
P75 = df_score["num_student"].quantile(75/100) # Percentile 75
P75
```

**3.2.7 พิสัย**

```python
#your code
range = df_score["num_student"].max() - df_score["num_student"].min() # พิสัย (range)
range
```

**3.2.8 ส่วนเบี่ยงเบนควอร์ไทล์**

```python
#your code
QD = Q3 - Q1 # ส่วนเบี่ยงเบนควอร์ไทล์ (Quartile deviation)
QD
```

**3.2.9 ส่วนเบี่ยงเบนมาตราฐานและความแปรปรวน**

```python
#your code
SD = df_score["num_student"].std() # ส่วนเบี่ยงเบนมาตราฐาน (Standard deviation)
VAR = df_score["num_student"].var() # ความแปรปรวน (Variance)
print(f"ส่วนเบี่ยงเบนมาตรฐาน = {SD}")
print(f"ความแปรปรวน = {VAR}") # SD ** 2
```

**3.2.10 สัมประสิทธิ์ของการแปรผัน**

```python
#your code
CV = (SD / mean) * 100 # สัมประสิทธิ์ของการแปรผัน (coefficient of variation)
CV
```

```python
FPS = [60.0, 30.2, 33.4, 59.9, 40.1, 25.5, 17.4, 28.6, 45.5, 57.8,
22.5, 50.2, 30.8, 43.1, 50.5, 35.5, 27.4, 29.2, 55.5, 10.2,
32.2, 27.8, 33.8, 53.4, 25.5, 30.4, 36.1, 49.3, 18.6, 50.0]
data = {"FPS": FPS}
df_fps = pd.DataFrame(data)
```

```python
#your code
FPS_mean = df_fps["FPS"].mean()
FPS_mean
```

ส่วนเบี่ยงเบนมาตราฐาน:

```python
#your code
FPS_SD = df_fps["FPS"].std()
FPS_SD
```

ฐานนิยม:

```python
#your code
FPS_mode = df_fps["FPS"].mode()
FPS_mode
```

Q<sub>1</sub>, D<sub>3</sub> และ P<sub>45</sub>:  

```python
#your code
FPS_Q1 = df_fps["FPS"].quantile(1/4)
FPS_D3 = df_fps["FPS"].quantile(3/10)
FPS_P45 = df_fps["FPS"].quantile(45/100)
print(f"Q1 = {FPS_Q1}")
print(f"D3 = {FPS_D3}")
print(f"P45 = {FPS_P45}")
```
